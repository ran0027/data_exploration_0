The purpose of this notebook is: to explore a .csv file.

```python
import csv
```

```python
with open("uab_fy2021_expenditures.csv") as f:
    uab_data = list(csv.DictReader(f))
    
print(uab_data[:5])
```

```python
print(uab_data[0])
```

```python
print(uab_data[0].keys())
```

```python
# Create a list of unique payees from uab_data transaction data

list_of_payees = []

for transaction in range(len(uab_data)):
    list_of_payees.append(uab_data[transaction]['Payee'])

unique_payees = list(set(list_of_payees))
print(unique_payees[:5])
```

```python
# Set up a nested dictionary which assigns a frequency of 0 and total amount paid of 0 to each unique payee
# IMPORTANT NOTE: Frequency refers to the number of times the payee shows up in the transactions list for FY 2021
# Total Amount Paid refers to the total amount of money that the payee was paid during FY 2021

dict_setup = {'Frequency': 0, 'Total Amount Paid': 0}

dict_setup_2 = [dict_setup for x in range(len(unique_payees))]

unique_payees_dict = dict(zip(unique_payees, dict_setup_2))

# Tests to figure out what went wrong with frequency & total amount paid when building unique payees dictionary

print(unique_payees_dict[unique_payees[20]].get('Frequency'))

print(unique_payees_dict[unique_payees[49]].get('Frequency'))

print(unique_payees_dict[unique_payees[20]].get('Total Amount Paid'))

print(unique_payees_dict[unique_payees[49]].get('Total Amount Paid'))
```

```python
# Iterate over list of transactions (in uab_data) to update frequency and total amount paid to each unique payee

for transaction in range(len(uab_data)):
    print(transaction)
    payee = uab_data[transaction]['Payee']
    print(payee)
    unique_payees_dict[payee]['Frequency'] = unique_payees_dict[payee].get('Frequency') + 1
    print(unique_payees_dict[payee]['Frequency']) # Something is wrong with frequency here; it is going up by 1 from payee to payee
    unique_payees_dict[payee]['Total Amount Paid'] = unique_payees_dict[payee].get('Total Amount Paid') + float(uab_data[transaction]['Amount'])
```

```python
# Print summary for one payee, with total amount paid and number of payments

print(f"{unique_payees[0].title()} was paid a total of ${unique_payees_dict[unique_payees[0]]['Total Amount Paid']} over the course of {unique_payees_dict[unique_payees[0]]['Frequency']} transactions in fiscal year 2021.")
```

```python
# Print summary for one payee, with total amount paid and number of payments

print(f"{unique_payees[1].title()} was paid a total of ${unique_payees_dict[unique_payees[1]]['Total Amount Paid']} over the course of {unique_payees_dict[unique_payees[1]]['Frequency']} transactions in fiscal year 2021.")
```

```python
for payee in unique_payees_dict:
    total = unique_payees_dict[payee]['Total Amount Paid']
    frequency = unique_payees_dict[payee]['Frequency']
    unique_payees_dict[payee]['Average Amount Per Transaction'] = total / frequency

print(unique_payees_dict[unique_payees[0]])
```

```python
# Sort unique payees dict by most frequently paid

most_frequently_paid = {k : v for k, v in sorted(unique_payees_dict.items(), key = lambda item : item[1]['Frequency'], reverse= True)}

# Sort unique payees dict by total amount paid

{k : v for k, v in sorted(unique_payees_dict.items(), key = lambda item : item[1]['Total Amount Paid'], reverse= True)}

# Sort unique payees dict by average amount per transaction

{k : v for k, v in sorted(unique_payees_dict.items(), key = lambda item : item[1]['Average Amount Paid'], reverse= True)}
```
